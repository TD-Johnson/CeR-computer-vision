## 1. Python Environment Setup

Before proceeding, ensure you have the following packages installed. Run the commands below:

**Option 1: Using Conda (Recommended)**
```bash
conda create -n cv-workshop python=3.10
conda activate cv-workshop
conda install pytorch torchvision pytorch-cuda=11.8 -c pytorch -c nvidia
pip install matplotlib pillow numpy jupyterlab
```

**Option 2: Using pip**
```bash
python -m venv cv-workshop
source cv-workshop/bin/activate  # On Windows: cv-workshop\Scripts\activate
pip install torch torchvision matplotlib pillow numpy jupyterlab
```

**Troubleshooting:**
- If CUDA is not needed, install PyTorch CPU version: `pip install torch torchvision`
- Check Python version: `python --version` (should be 3.8 or higher)

## 2. Verify Your Environment

Run the cell below to check your Python and PyTorch versions:

````python
import sys
print('Python version:', sys.version)

import torch
print('PyTorch version:', torch.__version__)
print('CUDA available:', torch.cuda.is_available())
if torch.cuda.is_available():
    print('GPU:', torch.cuda.get_device_name(0))
else:
    print('Running on CPU (this is fine for learning!)')
````
