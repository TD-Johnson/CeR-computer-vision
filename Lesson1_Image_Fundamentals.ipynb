## 1.1 Images as Arrays

Images are numerical arrays where each pixel has intensity values.
- **Grayscale image:** 2D array (height × width), values 0-255
- **Color image:** 3D array (height × width × channels), typically RGB with 3 channels

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Create a simple grayscale image (random noise)
gray_image = np.random.randint(0, 256, (100, 100), dtype=np.uint8)
print(f'Grayscale image shape: {gray_image.shape}')
print(f'Image dtype: {gray_image.dtype}')
print(f'Min pixel value: {gray_image.min()}, Max pixel value: {gray_image.max()}')

# Create a simple RGB image
rgb_image = np.random.randint(0, 256, (100, 100, 3), dtype=np.uint8)
print(f'\nRGB image shape: {rgb_image.shape}')
print(f'Number of channels: {rgb_image.shape[2]}')
```

## 1.2 Visualizing Images

Let's display both grayscale and RGB images:

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Display grayscale image
axes[0].imshow(gray_image, cmap='gray')
axes[0].set_title('Grayscale Image')
axes[0].axis('off')

# Display RGB image
axes[1].imshow(rgb_image)
axes[1].set_title('RGB Image')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

## 1.3 Loading Real Images

Load a sample image from torchvision datasets:

```python
import torchvision.transforms as transforms
from torchvision import datasets

# Download CIFAR-10
train_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transforms.ToTensor())

# Get a sample image
sample_image, label = train_dataset[0]
print(f'Image shape (channels, height, width): {sample_image.shape}')
print(f'Label: {label}')

# Convert tensor to numpy for visualization
sample_np = sample_image.numpy()
# Convert from CHW to HWC for matplotlib
sample_hw = np.transpose(sample_np, (1, 2, 0))
print(f'Reshaped for visualization (height, width, channels): {sample_hw.shape}')
```

## 1.4 Image Transforms

Transforms are operations applied to images for preprocessing and augmentation:

```python
# Define various transforms
transform_resize = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor()
])

transform_with_normalization = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# Load dataset with transforms
train_dataset_resized = datasets.CIFAR10(root='./data', train=True, download=False, transform=transform_resize)
img_resized, _ = train_dataset_resized[0]
print(f'Resized image shape: {img_resized.shape}')

# Visualize original and resized
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].imshow(np.transpose(sample_image.numpy(), (1, 2, 0)))
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(np.transpose(img_resized.numpy(), (1, 2, 0)))
axes[1].set_title('Resized to 64×64')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
